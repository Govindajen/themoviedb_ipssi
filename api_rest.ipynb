# Imports

```python
import os

import requests

import pandas as pd

import gzip
import json
from datetime import datetime

from dotenv import load_dotenv
```

# .gz

```python
url = 'https://files.tmdb.org/p/exports/movie_ids_10_17_2024.json.gz'

json_file = []

response = requests.get(url)
nom_fichier_gz = 'movie_ids_10_17_2024.json.gz'

with open(nom_fichier_gz, 'wb') as f:
    f.write(response.content)

with gzip.open(nom_fichier_gz, 'rt', encoding='utf-8') as f:
    for i, line in enumerate(f):
        try:
            json_file.append(json.loads(line))
        except:
            print("error")
        
        if i >= 999:  
            break

df = pd.DataFrame(json_file)

df
```
